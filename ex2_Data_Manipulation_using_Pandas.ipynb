```python
import numpy as np
import pandas as pd

# Load Dataset
dframe = pd.read_csv('Data.csv')  # Corrected quotes
print("Original DataFrame:")
print(dframe)

# Filling missing data with mean
dframe.fillna(value=dframe.mean(numeric_only=True), inplace=True)  # Avoids warnings for non-numeric columns
print("DataFrame after filling missing values:")
print(dframe)

# Dropping remaining missing data (if any non-numeric columns have NaNs)
dframe.dropna(inplace=True)
print("DataFrame after dropping remaining NaNs:")
print(dframe)
```

```python
import numpy as np

import pandas as pd
left = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'], 'A': ['A0', 'A1', 'A2', 'A3'],

'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'], 'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']})
# Merging the dataframes
print(pd.merge(left, right, how ='inner', on ='Key'))
# Joining the dataframes
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['B0', 'B1', 'B2', 'B3']}, index= ['K0', 'K1', 'K2', 'K3'])
right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']}, index= ['K0', 'K1', 'K2', 'K3'])
print(left.join(right))
```

```python
import numpy as np
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Adidas_Data.csv')  # Ensure the file is uploaded
print("Dataset Preview:\n", df.head())

# Descriptive statistics (numeric columns only)
print("\nMean:\n", df.mean(numeric_only=True))
print("\nCount:\n", df.count())
print("\nMedian:\n", df.median(numeric_only=True))
print("\nSum:\n", df.sum(numeric_only=True))

# ✅ Fix: Calculate Mean Absolute Deviation (MAD) manually
mad_values = df.select_dtypes(include=[np.number]).apply(lambda x: np.mean(np.abs(x - np.mean(x))))
print("\nMean Absolute Deviation (MAD):\n", mad_values)

print("\nStandard Deviation (STD):\n", df.std(numeric_only=True))
print("\nVariance:\n", df.var(numeric_only=True))
```
